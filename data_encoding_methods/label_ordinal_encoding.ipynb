# Label Encoding
Label and Ordinal encoding are 2 techniques to categorical data into numeric data.
Label encoding involves assigning a unique numerical label to each category in the variable. The labels are usually assigned in the alphabetical order or based on the frequency of their occurance.

For example, if we have a categorical variable color with 3 categories Red, Blue, Green, we can represent it as follows.
1. Red
2. Green
3. Blue

```python
import pandas as pd
df = pd.DataFrame(
    {'color': ['red', 'blue', 'green', 'green', 'red', 'blue']}
)
df
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
encoded_colors = lbl_encoder.fit_transform(df[['color']])
encoded_colors
```
