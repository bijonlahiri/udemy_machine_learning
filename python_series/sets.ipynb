```python
def merge_dicts_with_overlapping_keys(dicts):
    # Your code goes here
    merged_dict = {}
    for i in range(len(dicts)):
        for j in range(i+1, len(dicts)):
            for key in dicts[i].keys():
                if key in dicts[j].keys():
                    merged_dict[key] = dicts[i].get(key) + dicts[j].get(key)
                    print(f"Common Key --> {key}: {merged_dict.get(key)}")
                else:
                    merged_dict[key] = dicts[i].get(key)
                    print(f"Uncommon key --> {key}: {merged_dict.get(key)}")
    return merged_dict
            
dicts=[{'a':1,'b':2},{'b':3,'c':4},{'c':5,'d':6}]
print(merge_dicts_with_overlapping_keys(dicts))
```
