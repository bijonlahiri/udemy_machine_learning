```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris['data'], columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
df['target'] = iris['target']
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['target'], axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
tree_classifier = DecisionTreeClassifier(max_depth=2)
tree_classifier.fit(X_train, y_train)
```

```python
plt.figure(figsize=(15,10))
plot_tree(tree_classifier, filled=True)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, tree_classifier.predict(X_test)))
print(classification_report(y_test, tree_classifier.predict(X_test)))
```
