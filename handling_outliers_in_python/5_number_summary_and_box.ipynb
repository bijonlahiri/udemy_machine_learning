# 5 Number Summary and Box Plot
## Parameters
- Minimum
- Maximum
- Median
- Q1
- Q3
- IQR

```python
import numpy as np
lst_marks = [45, 32, 56, 75, 89, 54, 32, 89, 90, 87, 67, 54, 45, 98, 99, 67, 74]
minumum, Q1, median, Q3, maximum = np.quantile(lst_marks, [0.0, 0.25, 0.50, 0.75, 1.0])
```

```python
minumum, Q1, median, Q3, maximum
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_fence = Q1 - 1.5 * IQR
higher_fence = Q3 + 1.5 * IQR
```

```python
lower_fence, higher_fence
```

```python
import seaborn as sns
sns.boxplot(lst_marks)
```

```python
lst_marks = [-100, -120, -150, 45, 32, 56, 75, 89, 54, 32, 89, 90, 87, 67, 54, 45, 98, 99, 67, 74, 150, 170, 180]
sns.boxplot(lst_marks)
```
